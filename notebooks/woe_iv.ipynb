```python
from sklearn.datasets import load_breast_cancer
from mightypy.stats import WOE_IV

dataset = load_breast_cancer(as_frame=True)
df = dataset.frame[['mean radius', 'target']]

target_map = {
    0: 'False',
    1: 'True'
}

df['label'] = df['target'].map(target_map)

# print(df)
model = WOE_IV(event='True', non_event='False', target_col='label', bucket_col='mean radius')
```

```python
cal_df, iv = model.values(df=df)
```

```python
fig = model.plot(figsize=(15,7))
fig.tight_layout()
fig.show()
```
